## Understanding Context-Tree Weighting 
### Author: Gabriel Hofer  




```python
class ctw:
    def __init__(self,s,d):
        self.depth=d
        self.seq=s
        self.n=len(s)
        self.tree={}
        s=s[::-1]
        for i in range(len(s)-d):
            for j in range(1,d+1):
                if s[i:i+j] not in self.tree:
                    self.tree[s[i:i+j]]=int(0)
                self.tree[s[i:i+j]]+=int(1)
                
    def addBit(self,b):
        self.seq+=b
        for i in range(self.depth):
            if self.seq[self.n-i-1:] not in self.tree:
                self.tree[self.seq[self.n-i-1:]]=0
            self.tree[self.seq[self.n-i-1:]]+=int(1)
            
    def showTree(self):
        for i in self.tree:
            print(i+" --> "+str(self.tree[i]))
    
    def showSource(self):
        print(self.seq)
        
    def kt(self,a,b):
        if a: return ((a-0.5)/(a+b))*self.kt(a-1,b)
        elif b: return ((b-0.5)/(a+b))*self.kt(a,b-1)
        else: return 1
    
    def probs(self):
        for i in self.tree:
            [a,b]=self.getab(i)
            print(str(i)+"  a: "+str(a)+"  b: "+str(b)+"   "+str(self.kt(a,b)))
            
    def getab(self,s):
        a=b=0
        for i in s:
            if a: a+=1
            else: b+=1
        return [a,b]
        
```

```python
source = "1100100110"

t = ctw(source,3)
t.showTree()
t.probs()
#t.addBit('1')
#t.showTree()
```

## Understanding Context-Tree Weighting 
### Author: Gabriel Hofer  




[This](http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.109.1872&rep=rep1&type=pdf) article expalins context tree wieghting. Markdown guide: [here](https://www.markdownguide.org/basic-syntax/)

* Binary sequence of length T: $ x^T _1 = x_1 x_2 ... x_T $  
* A source code assigns to each source sequence $ x^T_1 $ a binary codeword $ c(x^T_1) $ with length $ L(x^T_1) $.
* $a_s$ and $b_s$ are the number of zeroes and ones in the subsequence associated with context $s$.
* $0s$ and $1s$ are the two children of $s$.
* $P^s_w$ is the weighted probability for $s$ with weight $w$.
* $\lambda$ denotes the root of the context tree which is an empty context.
* Each node $s$ in the context tree is associated with the subsequence of source symbols that occurred after context $s$.




Each node $s \in T_D$ with $l(s) \lt D $ "splits up" into two nodes, $0s$ and $1s$. 

Definition of _weighted context tree_:  
$ P_w^s = \begin{cases} \frac{1}{2}P_e(a_s,b_s) + \ \frac{1}{2}P_w^0s P_w^1s  \ \ \text{for} \ \ 0 \le l(s) \lt D \\ 
P_e(a_s,b_s). \ \ \text{for} \ \ l(s) = D \\ \end{cases}. $

The context tree together with the weighted probabilibies of the nodes is called a _weighted context tree_.

#### Binary Bounded Memory Tree Source Definition
A $binary \ tree \ source$ generates a sequence 
$x_{-\infty}^{\infty}$ 
of digits assuming values in teh alphabet ${0,1}$. We denote by 
$x_{m}^{n}$ the sequence $x_{m} x_{m+1} ... x_{n}$, and allow $n$ 
and $m$ to be infinitely large. For $n \lt m$ the sequence $x_{m}^{n}$ 
is empty, denoted by $\phi$.

# Look at the previous Bit !!

look at the bit before the current position/index to decide whether to put the index on the right or the left

Then, for the next (second) level or phase in the tree, look two bits to the left to determine which path to traverse.

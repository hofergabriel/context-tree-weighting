```python
class ctw:
    def __init__(self,s,d):
        self.seq=s
        self.depth=d
        self.n=len(s)
        self.tree={}
        s=s[::-1]
        self.tree['root']=[0,0]
        for i in range(len(s)-d):
            for j in range(0,d+1):
                if j==0: context='root' # root is the empty context
                else: context=s[i:i+j]
                
                if context not in self.tree: 
                    self.tree[context]=[0,0]
                if s[i+j:i+j+1]=='1': 
                    self.tree[context][1]+=int(1)
                else: 
                    self.tree[context][0]+=int(1)
                
                #print('\tcontext: '+context+"  "+str(self.tree[context]))
    
    # update context path
    def addBit(self,b):
        self.seq+=b
        self.n+=1
        if b: self.tree['root'][1]+=int(1)
        else: self.tree['root'][0]+=int(1)
        for i in range(1,self.depth+1):
            context=self.seq[self.n-i:]
            context=context[::-1]
            print("\tcontext: "+context)
            if context not in self.tree:
                self.tree[context]=[0,0]
            if self.seq[self.n-i-1:self.n-i]=='1':
                self.tree[context][1]+=int(1)
            else:
                self.tree[context][0]+=int(1)
            
    def showTree(self):
        for i in self.tree:
            print(i+" --> "+str(self.tree[i]))
    
    def showSource(self):
        print(self.seq)
        
    def kt(self,a,b):
        if a: return ((a-0.5)/(a+b))*self.kt(a-1,b)
        elif b: return ((b-0.5)/(a+b))*self.kt(a,b-1)
        else: return 1
    
    # TODO
    def probs(self):
        for i in self.tree:
            [a,b]=self.getab(i)
            print(str(i)+"    "+str(self.kt(a,b)))
            
    def getab(self,s):
        a=b=0
        for i in s:
            if a: a+=1
            else: b+=1
        return [a,b]
        
```

```python
source = "1100100110"

t = ctw(source,3)
t.showSource()
t.showTree()
#t.probs()
print()
t.addBit('1')
t.showSource()

t.showTree()
```
